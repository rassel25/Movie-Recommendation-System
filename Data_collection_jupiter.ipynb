```python
df = pd.DataFrame()
for i in range(1,506):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=ca218247dd7a4747395e5393caf26643&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id']]
    df = df.append(temp_df,ignore_index=True)
    
```

```python
for i in range(len(df)):
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key=ca218247dd7a4747395e5393caf26643&language=en-US'.format(df.loc[i,'id']))
    p= response.json()
    df.loc[i, ['genres']]=(", ".join([genre["name"] for genre in p["genres"]]))
```

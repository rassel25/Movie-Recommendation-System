```python
for i in range(1,506):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=ca218247dd7a4747395e5393caf26643&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview']]
    df = df.append(temp_df,ignore_index=True)
```

```python
df = pd.read_feather('id.feather')
```

```python
for i in range(len(df)):          # loop to collect generes
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key=ca218247dd7a4747395e5393caf26643&language=en-US'.format(df.loc[i,'id'])).json()
    df.loc[i, ['genres']]=(", ".join([genre["name"] for genre in response["genres"]]))
```

```python
df
```

```python
for i in range(len(df)):           # loop to collect keywords
    response = requests.get('https://api.themoviedb.org/3/movie/{}/keywords?api_key=ca218247dd7a4747395e5393caf26643'.format(df.loc[i,'id'])).json()
    df.loc[i, ['keywords']]=(", ".join([keywords["name"] for keywords in response["keywords"]]))
```

```python
df
```

```python
for i in range(len(df)):            # loop to collect upto 5 cast
    response = requests.get('https://api.themoviedb.org/3/movie/{}/credits?api_key=ca218247dd7a4747395e5393caf26643&language=en-US'.format(df.loc[i,'id'])).json()
    df.loc[i, ['cast']] = (", ".join([cast["name"] for cast in response["cast"][0:5]]))
```

```python
for i in range(len(df)):          # loop to collect Director
    response = requests.get('https://api.themoviedb.org/3/movie/{}/credits?api_key=ca218247dd7a4747395e5393caf26643&language=en-US'.format(df.loc[i,'id'])).json()
    temp_df = pd.DataFrame(response['crew'])
    df.loc[i, ['Director']] = (temp_df[temp_df["job"]=='Director'].groupby(['job']).agg({'name': lambda x: ", ".join(x)},axis=1))['name'].values        # .values is used to take the name without index
    
```
